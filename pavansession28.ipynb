```python
title = soup.find("div", attrs={"class":"_3wU53n"})
print(title.text)
```

```python
titles = soup.find_all("div", attrs={"class":"_3wU53n"})
for title in titles:
    print(title.text)
```

```python
cards = soup.find("div",attrs={"class":"_1UoZlX"})
title = card.find("div", attrs={"class":"_3wU53n"})
price = card.find("div",attrs={"class":"_2rQ-NK"})
rating = card.find("div",attrs={"class":"hGSR34"})
print("{} {} {}".format(title.text,price.text,rating.text))
```

```python
HOME_URL = "https://www.flipkart.com"
next_link = soup.find("a",attrs={"class":"_3fVaIS"})
print(HOME_URL)
response2 = requests.get(HOME_URL)
print(response2.status_code)
soup2 = BeautifulSoup(response2.content,"html.parser")
cards = soup2.find_all("div",attrs={"class":"_1UoZlX"})
for card in cards:
    title = card.find("div", attrs={"class":"_3wU53n"})
    price = card.find("div",attrs={"class":"_2rQ-NK"})
    rating = card.find("div",attrs={"class":"hGSR34"})
    print("{} {} {}".format(title.text,price.text,rating.text))
```

```python
HOME_URL = "https://www.flipkart.com"
next_link = soup.find("a",attrs={"class":"_3fVaIS"})
print(HOME_URL+next_link.get("href"))

response2 = requests.get(HOME_URL+next_link.get("href"))
print(response2.status_code)
soup2 = BeautifulSoup(response2.content,"html.parser")
cards = soup2.find_all("div",attrs={"class":"_1UoZlX"})
for card in cards:
    title = card.find("div", attrs={"class":"_3wU53n"})
    price = card.find("div",attrs={"class":"_2rQ-NK"})
    rating = card.find("div",attrs={"class":"hGSR34"})
    print("{} {} {}".format(title.text,price.text,rating.text))
```
